```python
import numpy as np
import matplotlib.pyplot as plt
import os

# -----------------------------
# Fonctions utilitaires
# -----------------------------
def read_params(file):
    params = {}
    with open(file, "r", encoding="utf-8") as f:
        first_line = f.readline().strip()
        for part in first_line.split(";"):
            sub = part.strip().split()
            if len(sub) >= 2:
                key, val = sub[0], sub[1]
                try:
                    params[key] = float(val) if "." in val else int(val)
                except:
                    params[key] = val
    return params

def find_value(file, key):
    with open(file, "r", encoding="utf-8") as f:
        for line in f:
            if key in line:
                return line
    return None

def extract_value(line, key):
    if line is None:
        return None
    parts = line.replace(";", " ").split()
    for i in range(len(parts)-1):
        if parts[i] == key:
            try:
                return float(parts[i+1])
            except:
                return None
    return None

# -----------------------------
# Fichiers
# -----------------------------
file = "data/001MoDe_R1.csv"
file2 = "data/001MoDe_R1.marker.csv"

params = read_params(file)
CENTER_X = params["centerX"]
CENTER_Y = params["centerY"]
cursor_radius = params["cursorRadius"]
internal_radius = params["internalRadius"] + cursor_radius
external_radius = params["externalRadius"] - cursor_radius

SCREEN_W = extract_value(find_value(file2,"screenWidth"),"screenWidth")
SCREEN_H = extract_value(find_value(file2,"screenHeight"),"screenHeight")

# -----------------------------
# Chargement des donnes
# -----------------------------
data = np.genfromtxt(file, delimiter=",", skip_header=3, names=True)
ts = data["timestamp"]
mx_raw = data["mouseX"]
my_raw = data["mouseY"]
mt = data["mouseInTarget"]

# -----------------------------
# Segmentation
# -----------------------------
segments = []
start = None
for i in range(len(ts)):
    if ts[i] > 0 and start is None:
        start = i
    if ts[i] == 0 and start is not None:
        segments.append((start,i))
        start = None
if start is not None:
    segments.append((start,len(ts)))

# -----------------------------
# Fonction de trac d'un segment
# -----------------------------
def plot_segment(num, i0, i1):
    T = (ts[i0:i1] - ts[i0]) / 1000
    X = mx_raw[i0:i1] - CENTER_X
    Y = -(my_raw[i0:i1] - CENTER_Y)   # inversion Y
    MT = mt[i0:i1]

    dist = np.sqrt(X**2 + Y**2)
    inside = (dist >= internal_radius) & (dist <= external_radius)
    outside = ~inside

    fig = plt.figure(figsize=(12,16))
    gs = fig.add_gridspec(7,1)

    ax0 = fig.add_subplot(gs[0:3,0])
    ax1 = fig.add_subplot(gs[3,0])
    ax2 = fig.add_subplot(gs[4,0])
    ax3 = fig.add_subplot(gs[5:,0])

    # --- Anneau jaune ---
    theta = np.linspace(0,2*np.pi,600)
    xo = external_radius*np.cos(theta)
    yo = external_radius*np.sin(theta)
    xi = internal_radius*np.cos(theta)
    yi = internal_radius*np.sin(theta)
    ax0.fill(xo,yo,color="yellow")
    ax0.fill(xi,yi,color="white")   # trou central

    # --- Cadre rose centr ---
    halfW = SCREEN_W/2
    halfH = SCREEN_H/2
    frame_x = [-halfW, halfW, halfW, -halfW, -halfW]
    frame_y = [-halfH, -halfH, halfH, halfH, -halfH]
    ax0.plot(frame_x, frame_y, color="magenta", linewidth=2)

    # --- Points verts/rouges ---
    ax0.scatter(X[inside],Y[inside],c="green",s=4)
    ax0.scatter(X[outside],Y[outside],c="red",s=4)
    ax0.plot(X,Y,color="gray",linewidth=0.5)

    ax0.set_xlim(-600,600)
    ax0.set_ylim(-400,400)
    ax0.set_aspect("equal")
    ax0.grid(True,alpha=0.3)
    ax0.set_title(f"Segment {num}")

    # --- Graphiques temporels ---
    ax1.plot(T,X,marker=".",markersize=3)
    ax1.grid(True); ax1.set_ylabel("X")

    ax2.plot(T,Y,marker=".",markersize=3,color="orange")
    ax2.grid(True); ax2.set_ylabel("Y (invers)")

    ax3.plot(T,MT,marker=".",markersize=3,color="green")
    ax3.set_ylim(-0.1,1.1); ax3.grid(True); ax3.set_ylabel("in target")

    plt.tight_layout()
    plt.show()

# -----------------------------
# Trac de tous les segments
# -----------------------------
for idx,(s,e) in enumerate(segments):
    plot_segment(idx,s,e)
```

```python
SEPARATOR = ';' 
HEADER_COLUMNS = "timestamp,mouseX,mouseY,mouseInTarget"

def find_dynamic_skip_header(filepath, header_line_start):
    skip_count = 0
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            for line in f:
                stripped_line = line.strip()
                if stripped_line.startswith(header_line_start):
                    return skip_count 
                
                if stripped_line:
                    skip_count += 1
                
                if skip_count > 100:
                    break 
            
    except FileNotFoundError:
        return -1
        
    return -1


DYNAMIC_SKIP_HEADER = find_dynamic_skip_header(OurFile, HEADER_COLUMNS)

if DYNAMIC_SKIP_HEADER == -1:
    print(f"Error: The head line '{HEADER_COLUMNS}' has not been found in the file '{OurFile}'. Check the format.")
    exit()

DYNAMIC_SKIP_DATA = DYNAMIC_SKIP_HEADER + 1 

print(f"Header lines detected: {DYNAMIC_SKIP_HEADER}. Data will start after {DYNAMIC_SKIP_DATA} lines.")


with open(OurFile, 'r', encoding='utf-8') as f:
    firstLine = f.readline().strip()

if firstLine:
    data_header = firstLine.split(';')
    print(f"Header line information in '{OurFile}':")
    for d in data_header:
        print("  " + d.strip())

print(f"Markers in '{OurFile2}':")
print(OurFile)

lines_after_date = []
processing_block = False

def check_date_format_manual(s):
    if len(s) < 10:
        return False
    
    if s[4] != '-' or s[7] != '-':
        return False
    
    try:
        if not s[0:4].isdigit(): return False
        if not s[5:7].isdigit(): return False
        if not s[8:10].isdigit(): return False
        return True
    except IndexError:
        return False

def print_aligned_block(lines):
    if not lines:
        return

    split_lines = [line.split(SEPARATOR) for line in lines] 

    max_widths = []
    num_cols = max(len(cols) for cols in split_lines) if split_lines else 0

    for col in range(num_cols):
        max_width = 0
        for cols in split_lines:
            if col < len(cols):
                max_width = max(max_width, len(cols[col].strip()))
        max_widths.append(max_width)

    for cols in split_lines:
        padded = []
        for i, col in enumerate(cols):
            content = col.strip()
            padded.append(content.ljust(max_widths[i]))
        print(" " + " , ".join(padded))

try:
    with open(OurFile2, 'r', encoding='utf-8') as f:
        for i in range(3):
            next(f) 
        
        for line in f:
            stripped = line.strip()

            if check_date_format_manual(stripped):
                
                print_aligned_block(lines_after_date)
                lines_after_date = []
                processing_block = True

                parts = stripped.split(SEPARATOR, 1)
                stripped_content = parts[1].strip() if len(parts) > 1 else stripped

                subparts = stripped_content.split(SEPARATOR, 1)
                if len(subparts) == 2:
                    stripped_output = f"{subparts[0].strip()}: {subparts[1].strip()}"
                else:
                    stripped_output = stripped_content

                print(stripped_output)

            elif processing_block:
                lines_after_date.append(stripped)
            
            else:
                print(" " + stripped)

except FileNotFoundError:
    print(f"Error: The file '{OurFile2}' is unfindable. Please choose another path.")
except Exception as e:
    print(f"An error has happened during the file's treatment: {e}")

print_aligned_block(lines_after_date)


def get_column(OurFile, col="A", sep=";"):
    col_index = 0
    for char in col.upper():
        col_index = col_index * 26 + (ord(char) - ord("A") + 1)
    col_index -= 1

    values = []
    with open(OurFile, "r", encoding="utf-8") as f:
        lines = f.readlines()

    for line in lines[DYNAMIC_SKIP_DATA:]:
        cols = line.strip().split(sep)[0].split(',') 
        if col_index < len(cols):
            values.append(cols[col_index])
        else:
            values.append("")
    return values


def find_segments_from_column(col):
    segments = []
    start = None

    if col and col[0] != "0":
        start = 0

    for i in range(1, len(col)):
        curr = col[i]
        prev = col[i-1]

        if prev == "0" and curr != "0" and start is None:
            start = i

        if curr == "0" and start is not None:
            segments.append((start, i - 1))
            start = None

    if start is not None:
        segments.append((start, len(col) - 1))

    return segments


def compute_segment_times(col, segments):
    times = []
    for start, end in segments:
        try:
            t_start = int(col[start])
            t_end = int(col[end])
            duration = (t_end - t_start) / 1000.0
            times.append((start, end, t_start, t_end, duration))
        except ValueError:
            continue
    return times

def sanitize_column(col):
    clean = []
    for val in col:
        try:
            int(val)
            clean.append(val)
        except:
            clean.append("0") 
    return clean

colA_raw = get_column(OurFile, "A", sep=SEPARATOR)
colA = sanitize_column(colA_raw)


segments = find_segments_from_column(colA)
results = compute_segment_times(colA, segments)

print("record times (milliseconds since the Unix epoch, 00:00:00 UTC on January 1, 1970):")
for i, (s, e, t1, t2, dur) in enumerate(results, 1):
    print(f" {t1} to {t2} ({dur:.3f} s)")
    
print("\n")

if len(results) > 0:
    first_dorecord_marker = int(results[0][2])
else:
    first_dorecord_marker = 0
print("record times (seconds since the first DoRecord marker):")
for i, (s, e, t_start, t_end, dur) in enumerate(results, 1):
    t_start_sec = (int(t_start) - first_dorecord_marker) / 1000
    t_end_sec = (int(t_end) - first_dorecord_marker) / 1000
    print(f"{t_start_sec:.3f} to {t_end_sec:.3f}({dur:.3f} s)")

def find_value_in_csv(OurFile, key, sep=";"):
    with open(OurFile, "r", encoding="utf-8") as f:
        for i, line in enumerate(f):
            if i >= DYNAMIC_SKIP_HEADER:
                break
            cells = line.strip().split(sep)
            for cell in cells:
                if key in cell:
                    return cell
    return None

def extract_numeric_value(cell, key):
    if cell is None:
        return None
    parts = cell.replace(";", " ").split() 
    for i in range(len(parts) - 1):
        if parts[i] == key:
            try:
                return float(parts[i + 1])
            except:
                return None
    return None

raw_centerX = find_value_in_csv(OurFile, "centerX", sep=SEPARATOR)
raw_centerY = find_value_in_csv(OurFile, "centerY", sep=SEPARATOR)

CENTER_X = extract_numeric_value(raw_centerX, "centerX")
CENTER_Y = extract_numeric_value(raw_centerY, "centerY")

timestamps_list = []
mouseX_list = []
mouseY_list = []
mouseInTarget_list = []

with open(OurFile, 'r') as f:
    lines = f.readlines()
    for line in lines[DYNAMIC_SKIP_DATA:]:
        line = line.strip()
        if not line:
            continue
        
        data_block = line.split(SEPARATOR, 1)[0] 
        
        parts = data_block.split(',') 
        
        if len(parts) >= 4:
            try:
                ts = int(parts[0])
                mx = float(parts[1])
                my = float(parts[2])
                mt = int(parts[3])
                
                timestamps_list.append(ts)
                mouseX_list.append(mx)
                mouseY_list.append(my)
                mouseInTarget_list.append(mt)
            except:
                continue

if not timestamps_list:
    print("WARNING: No data read for the NumPy/Plotting treatment.")
    exit()
else:
    timestamp = np.array(timestamps_list, dtype=np.int64)
    mouseX = np.array(mouseX_list, dtype=float)
    mouseY = np.array(mouseY_list, dtype=float)
    mouseInTarget = np.array(mouseInTarget_list, dtype=int) 

    mouseX_centered = mouseX - CENTER_X
    mouseY_centered = mouseY - CENTER_Y

    def find_segments(ts):
        segments = []
        start = None
        if ts[0] > 0:
            start = 0
        for i in range(1, len(ts)):
            if ts[i-1] == 0 and ts[i] > 0:
                start = i
            if ts[i] == 0 and start is not None:
                segments.append((start, i))
                start = None
        if start is not None:
            segments.append((start, len(ts)))
        return segments

    segments = find_segments(timestamp)

    def extract_segment(start_idx, end_idx):
        return (
            timestamp[start_idx:end_idx],
            mouseX_centered[start_idx:end_idx],
            mouseY_centered[start_idx:end_idx]
        )

    plt.figure(figsize=(12, 6))

    first = True
    for i, (start, end) in enumerate(segments):
        ts_seg, x_seg, y_seg = extract_segment(start, end)
        ts_seg = np.where(ts_seg == 0, np.nan, ts_seg)
        t_rel_global = (ts_seg - timestamp[0]) / 1000.0
        if first:
            plt.plot(t_rel_global, x_seg, color="tab:blue", linewidth=1.5, label="x position (centered)")
            plt.plot(t_rel_global, y_seg, color="tab:orange", linewidth=1.5, label="y position (centered)")
            first = False
        else:
            plt.plot(t_rel_global, x_seg, color="tab:blue", linewidth=1.5)
            plt.plot(t_rel_global, y_seg, color="tab:orange", linewidth=1.5)

    valid_ts_indices = np.where(timestamp != 0)[0]
    if len(valid_ts_indices) > 0:
        first_valid_idx = valid_ts_indices[0]
        t_first_valid = (timestamp[first_valid_idx] - timestamp[0]) / 1000.0
        plt.axvline(x=t_first_valid, color='black', linestyle='--', linewidth=1)

    zero_indices = np.where(timestamp == 0)[0]

    record_start_added = False
    record_end_added = False

    for idx in zero_indices:
        prev_idx = idx - 1
        next_idx = idx + 1
        if prev_idx >= 0 and timestamp[prev_idx] != 0:
            t_black = (timestamp[prev_idx] - timestamp[0]) / 1000.0
            if not record_end_added:
                plt.axvline(x=t_black, color='red', linestyle='--', linewidth=1, label="record end")
                record_end_added = True
            else:
                plt.axvline(x=t_black, color='red', linestyle='--', linewidth=1)
        if next_idx < len(timestamp) and timestamp[next_idx] != 0:
            t_red = (timestamp[next_idx] - timestamp[0]) / 1000.0
            if not record_start_added:
                plt.axvline(x=t_red, color='black', linestyle='--', linewidth=1, label="record start")
                record_start_added = True
            else:
                plt.axvline(x=t_red, color='black', linestyle='--', linewidth=1)

    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title('Raw time series data (dynamic segmentation)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

valid_mask = ~((mouseX == 0) & (mouseY == 0))
mouseX_filtered = mouseX[valid_mask]
mouseY_filtered = mouseY[valid_mask]
mouseInTarget_filtered = mouseInTarget[valid_mask]

if len(mouseX_filtered) == 0:
    print("ERROR : Ther's no more data to print after filtration. Verify the file or the mask.")
    exit()

mean_x = np.mean(mouseX_filtered)
mean_y = np.mean(mouseY_filtered)

mouseX_centered_plot = mouseX_filtered - mean_x
mouseY_centered_plot = mouseY_filtered - mean_y

fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(mouseX_centered_plot, mouseY_centered_plot, 'b-', linewidth=1.5, label='trajectory')

inside_mask = mouseInTarget_filtered == 1

if np.any(inside_mask):
    center_x = np.mean(mouseX_centered_plot[inside_mask])
    center_y = np.mean(mouseY_centered_plot[inside_mask])

    distances = np.sqrt((mouseX_centered_plot[inside_mask] - center_x)**2 + 
                        (mouseY_centered_plot[inside_mask] - center_y)**2)
    outer_radius = np.max(distances)
    inner_radius = np.min(distances)
    
    circle_outer = plt.Circle((center_x, center_y), outer_radius, 
                              color='red', fill=False, linewidth=2, label='target')
    circle_inner = plt.Circle((center_x, center_y), inner_radius, 
                              color='red', fill=False, linewidth=2)
    ax.add_patch(circle_outer)
    ax.add_patch(circle_inner)

    ax.scatter(mouseX_centered_plot[inside_mask], mouseY_centered_plot[inside_mask], 
               c='green', s=30, alpha=0.7, label='inside target', zorder=3)
else:
    print("WARNING: No data has mouseInTarget = 1 after filtration. Circles won't be printed.")
    center_x = np.mean(mouseX_centered_plot) if len(mouseX_centered_plot) > 0 else 0
    center_y = np.mean(mouseY_centered_plot) if len(mouseY_centered_plot) > 0 else 0


ax.set_xlabel('x position (Centered)', fontsize=12)
ax.set_ylabel('y position (Centered)', fontsize=12)
ax.set_title('Centered Trajectory', fontsize=14)
ax.legend(loc='upper right', fontsize=10)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')

plt.tight_layout()
plt.show()

timestamps_stat = timestamp
mouseX_stat = mouseX
mouseY_stat = mouseY
mouseInTarget_stat = mouseInTarget

zero_mask = (mouseX_stat == 0) & (mouseY_stat == 0)
zero_indices = np.where(zero_mask)[0]

records = []
start_idx = 0

for zero_idx in zero_indices:
    if zero_idx > start_idx:
        record_data = {
            'timestamps': timestamps_stat[start_idx:zero_idx],
            'mouseX': mouseX_stat[start_idx:zero_idx],
            'mouseY': mouseY_stat[start_idx:zero_idx],
            'mouseInTarget': mouseInTarget_stat[start_idx:zero_idx]
        }
        records.append(record_data)
    start_idx = zero_idx + 1

if start_idx < len(mouseX_stat):
    record_data = {
        'timestamps': timestamps_stat[start_idx:],
        'mouseX': mouseX_stat[start_idx:],
        'mouseY': mouseY_stat[start_idx:],
        'mouseInTarget': mouseInTarget_stat[start_idx:]
    }
    records.append(record_data)

t_reference = None
for record in records:
    valid_timestamps = record['timestamps'][~np.isnan(record['timestamps'])]
    if len(valid_timestamps) > 0:
        t_reference = valid_timestamps[0]
        break

print(f"\nNumber of records: {len(records)}")
for i, record in enumerate(records):
    valid_mask = ~np.isnan(record['timestamps'])
    valid_timestamps = record['timestamps'][valid_mask]
    
    n_samples = len(valid_timestamps)
    if n_samples > 0:
        t_start = (valid_timestamps[0] - t_reference) / 1000 if t_reference is not None else 0
        t_end = (valid_timestamps[-1] - t_reference) / 1000 if t_reference is not None else 0
        duration = (valid_timestamps[-1] - valid_timestamps[0]) / 1000
        frequency = n_samples / duration if duration > 0 else 0
        
        print(f"record {i}: {n_samples} samples from {t_start:.3f} to {t_end:.3f} s, ({duration:.3f}s, {frequency:.3f} Hz)")
    else:
        print(f"Record {i}: No valid timestamp")

def read_params(OurFile):
    params = {}
    with open(OurFile, "r", encoding="utf-8") as f:
        first_line = f.readline().strip()
        for part in first_line.split(";"):
            sub = part.strip().split()
            if len(sub) >= 2:
                key, val = sub[0], sub[1]
                try:
                    params[key] = float(val) if "." in val else int(val)
                except:
                    params[key] = val
    return params

def find_value(OurFile, key):
    with open(OurFile, "r", encoding="utf-8") as f:
        for i, line in enumerate(f):
            if i >= DYNAMIC_SKIP_HEADER:
                break
            if key in line:
                return line
    return None

def extract_value(line, key):
    if line is None:
        return None
    parts = line.replace(";", " ").split()
    for i in range(len(parts)-1):
        if parts[i] == key:
            try:
                return float(parts[i+1])
            except:
                return None
    return None


params = read_params(OurFile)
CENTER_X = params.get("centerX")
CENTER_Y = params.get("centerY")
cursor_radius = params.get("cursorRadius", 0)
internal_radius = params.get("internalRadius", 0) + cursor_radius
external_radius = params.get("externalRadius", 0) - cursor_radius

SCREEN_W = extract_value(find_value(OurFile2,"screenWidth"),"screenWidth")
SCREEN_H = extract_value(find_value(OurFile2,"screenHeight"),"screenHeight")

ts = timestamp
mx_raw = mouseX
my_raw = mouseY
mt = mouseInTarget

segments = []
start = None
for i in range(len(ts)):
    if ts[i] > 0 and start is None:
        start = i
    if ts[i] == 0 and start is not None:
        segments.append((start,i))
        start = None
if start is not None:
    segments.append((start,len(ts)))

def plot_segment(num, i0, i1):
    T = (ts[i0:i1] - ts[i0]) / 1000
    X = mx_raw[i0:i1] - CENTER_X
    Y = -(my_raw[i0:i1] - CENTER_Y) 
    MT = mt[i0:i1]

    dist = np.sqrt(X**2 + Y**2)
    
    inside = (dist >= internal_radius) & (dist <= external_radius)
    outside = ~inside

    fig = plt.figure(figsize=(12,16))
    gs = fig.add_gridspec(7,1)

    ax0 = fig.add_subplot(gs[0:3,0])
    ax1 = fig.add_subplot(gs[3,0])
    ax2 = fig.add_subplot(gs[4,0])
    ax3 = fig.add_subplot(gs[5:,0])

    theta = np.linspace(0,2*np.pi,600)
    xo = external_radius*np.cos(theta)
    yo = external_radius*np.sin(theta)
    xi = internal_radius*np.cos(theta)
    yi = internal_radius*np.sin(theta)
    ax0.fill(xo,yo,color="yellow")
    ax0.fill(xi,yi,color="white")

    halfW = SCREEN_W/2 if SCREEN_W is not None else 800/2
    halfH = SCREEN_H/2 if SCREEN_H is not None else 600/2
    frame_x = [-halfW, halfW, halfW, -halfW, -halfW]
    frame_y = [-halfH, -halfH, halfH, halfH, -halfH]
    ax0.plot(frame_x, frame_y, color="magenta", linewidth=2)

    ax0.scatter(X[inside],Y[inside],c="green",s=4)
    ax0.scatter(X[outside],Y[outside],c="red",s=4)
    ax0.plot(X,Y,color="gray",linewidth=0.5)

    ax0.set_xlim(-(halfW + 100), halfW + 100)
    ax0.set_ylim(-(halfH + 100), halfH + 100)
    ax0.set_aspect("equal")
    ax0.grid(True,alpha=0.3)
    ax0.set_title(f"Segment {num}")

    ax1.plot(T,X,marker=".",markersize=3)
    ax1.grid(True); ax1.set_ylabel("X")

    ax2.plot(T,Y,marker=".",markersize=3,color="orange")
    ax2.grid(True); ax2.set_ylabel("Y (reversed)")

    ax3.plot(T,MT,marker=".",markersize=3,color="green")
    ax3.set_ylim(-0.1,1.1); ax3.grid(True); ax3.set_ylabel("in target")

    plt.tight_layout()
    plt.show()

for idx,(s,e) in enumerate(segments):
    plot_segment(idx,s,e)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# =================================================================
# 1. DFINITION DES FICHIERS ET SPARATEURS
# =================================================================
# Fichiers
DATA_FILE = "data/data.csv"       # <-- IMPORTANT : chemin du fichier
df_data = pd.read_csv(DATA_FILE, skiprows=3)   # <-- DataFrame charg correctement

MARKER_FILE = "data/marker.csv"   # NotreFile2 (Marqueurs d'vnement) 

# Sparateurs
SEPARATOR_PARAM = ';' 
SEPARATOR_DATA = ',' 
HEADER_COLUMNS = "timestamp,mouseX,mouseY,mouseInTarget"


# =================================================================
# 2 & 3. DTECTION DU HEADER & PARAMTRES (Code omis pour concision)
# =================================================================

def find_dynamic_skip_header(filepath, header_line_start):
    skip_count = 0
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            for line in f:
                stripped_line = line.strip()
                if stripped_line.startswith(header_line_start): 
                    return skip_count 
                if stripped_line: 
                    skip_count += 1
                if skip_count > 100: 
                    break 
    except FileNotFoundError: 
        return -1
    return -1


def extract_center_values(filepath, key):
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            first_line = f.readline().strip()
            parts = first_line.split(SEPARATOR_PARAM)
            for part in parts:
                part = part.strip()
                if part.startswith(key):
                    sub_parts = part.split()
                    if len(sub_parts) == 2:
                        try: 
                            return float(sub_parts[1])
                        except ValueError: 
                            return None
    except FileNotFoundError: 
        return None
    return None


DYNAMIC_SKIP_HEADER = find_dynamic_skip_header(DATA_FILE, HEADER_COLUMNS)
if DYNAMIC_SKIP_HEADER == -1:
    exit()

DYNAMIC_SKIP_DATA = DYNAMIC_SKIP_HEADER + 1 
print(f"Les donnes commencent aprs {DYNAMIC_SKIP_DATA} lignes dans {DATA_FILE}.")

CENTER_X = extract_center_values(DATA_FILE, "centerX") or 0
CENTER_Y = extract_center_values(DATA_FILE, "centerY") or 0
print(f"Centrage utilis: CENTER_X={CENTER_X}, CENTER_Y={CENTER_Y}")


# =================================================================
# 4. CHARGEMENT DES DONNES DE TRAJECTOIRE (DATA_FILE)
# =================================================================

timestamps_list = []
mouseX_list = []
mouseY_list = []

try:
    with open(DATA_FILE, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines[DYNAMIC_SKIP_DATA:]:
            line = line.strip()
            if not line:
                continue
            
            data_block = line.split(SEPARATOR_PARAM, 1)[0]
            parts = data_block.split(SEPARATOR_DATA)
            
            if len(parts) >= 3:
                try:
                    ts = int(parts[0].strip()) if parts[0].strip().isdigit() else 0
                    mx = float(parts[1].strip()) if parts[1].strip() else 0.0
                    my = float(parts[2].strip()) if parts[2].strip() else 0.0
                    
                    timestamps_list.append(ts)
                    mouseX_list.append(mx)
                    mouseY_list.append(my)
                except Exception:
                    continue
except FileNotFoundError:
    print(f"ERREUR FATALE: Le fichier de donnes {DATA_FILE} est introuvable.")
    exit()

timestamp = np.array(timestamps_list, dtype=np.int64)
mouseX = np.array(mouseX_list)
mouseY = np.array(mouseY_list)

if len(timestamp) == 0:
    print("ERREUR: Aucune donne valide n'a t charge.")
    exit()


# =================================================================
# 5. CHARGEMENT ET EXTRACTION DES MARQUEURS (MARKER_FILE) - CORRIG
# =================================================================

ts_do_record = []
ts_do_pause = []

try:
    with open(MARKER_FILE, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = [p.strip() for p in line.split(',')]

            if len(parts) >= 3 and parts[1].isdigit():
                epoch = int(parts[1])
                marker_text = " ".join(parts[2:])

                if "DoRecord" in marker_text:
                    ts_do_record.append(epoch)

                elif "DoPause" in marker_text:
                    ts_do_pause.append(epoch)

    ts_do_record = np.array(ts_do_record, dtype=np.int64)
    ts_do_pause = np.array(ts_do_pause, dtype=np.int64)

    print(f"\nDiagnostic Marqueurs : {len(ts_do_record)} DoRecord trouvs. {len(ts_do_pause)} DoPause trouvs.")

except FileNotFoundError:
    print(f"\nERREUR: Le fichier marqueur {MARKER_FILE} est introuvable. Segmentation impossible.")
    ts_do_record = np.array([])
    ts_do_pause = np.array([])


# =================================================================
# 6. SEGMENTATION PAR MARQUEURS (OU FALLBACK)
# =================================================================

if len(ts_do_record) < 1 or len(ts_do_pause) < 1:
    print("\n------------------------------------------------------------------------------------------------------")
    print("CHEC DE SEGMENTATION PAR MARQUEURS. Retour  la mthode fiable (transition timestamp 0 <-> non-0).")
    print("------------------------------------------------------------------------------------------------------")

    def find_segments_fallback(ts):
        segments = []
        start = None
        if len(ts) > 0 and ts[0] > 0:
            start = 0

        for i in range(1, len(ts)):
            if ts[i-1] == 0 and ts[i] > 0:
                start = i
            if ts[i] == 0 and start is not None:
                segments.append((start, i - 1))
                start = None

        if start is not None:
            segments.append((start, len(ts) - 1))
        return segments

    segments = find_segments_fallback(timestamp)

else:
    segments = []
    num_records = min(len(ts_do_record), len(ts_do_pause))

    print(f"\n{num_records} records dtects par les marqueurs externes.")

    for i in range(num_records):
        ts_start = ts_do_record[i]
        ts_end = ts_do_pause[i]

        start_index = np.searchsorted(timestamp, ts_start, side='left')
        end_index = np.searchsorted(timestamp, ts_end, side='right') - 1

        if start_index < len(timestamp) and end_index >= start_index:
            segments.append((start_index, end_index))


# =================================================================
# 7. STATISTIQUES (IDENTIQUE)
# =================================================================

first_non_zero_ts_index = np.where(timestamp != 0)[0]
if len(first_non_zero_ts_index) > 0:
    t_ref = timestamp[first_non_zero_ts_index[0]]
else:
    t_ref = 0

print("\nAttribution des Timestamps (ms) et Dures (s) pour chaque record dtect:")
print("--------------------------------------------------------------------------")

records_found = 0
for i, (s, e) in enumerate(segments):
    segment_timestamps = timestamp[s:e+1]
    
    valid_mask = segment_timestamps != 0
    valid_segment_timestamps = segment_timestamps[valid_mask]
    
    n_samples = len(valid_segment_timestamps)
    
    if n_samples > 0:
        records_found += 1
        
        ts_abs_start = valid_segment_timestamps[0]
        ts_abs_end = valid_segment_timestamps[-1]
        
        t_start_rel = (ts_abs_start - t_ref) / 1000.0
        t_end_rel = (ts_abs_end - t_ref) / 1000.0
        duration = (ts_abs_end - ts_abs_start) / 1000.0
        
        print(f"Record {i} - Dbut : {t_start_rel:.3f} s (TS Absolu: {ts_abs_start})")
        print(f"Record {i} - Fin : {t_end_rel:.3f} s (TS Absolu: {ts_abs_end})")
        print(f"Duration : {duration:.3f} s | Samples : {n_samples}")
        print(f"Indices de lignes dans {DATA_FILE} : {s}  {e}")
        print("---")
    else:
        print(f"Record {i}: Aucune donne valide dans [{s}, {e}].")

if records_found == 0:
    print("Aucun record valide trouv.")


# =================================================================
# 8. TRAC (IDENTIQUE)
# =================================================================

mouseX_centered = mouseX - CENTER_X
mouseY_centered = mouseY - CENTER_Y
inactive_mask = (timestamp == 0)

x_plot = mouseX_centered.copy()
y_plot = mouseY_centered.copy()

x_plot[inactive_mask] = np.nan
y_plot[inactive_mask] = np.nan

plt.figure(figsize=(12, 6))
t_rel_global = (timestamp - t_ref) / 1000.0

plt.plot(t_rel_global, x_plot, color="tab:blue", linewidth=1.5, label="x position (centered)")
plt.plot(t_rel_global, y_plot, color="tab:orange", linewidth=1.5, label="y position (centered)")

for i, (start_idx, end_idx) in enumerate(segments):
    t_start_record = (timestamp[start_idx] - t_ref) / 1000.0
    t_end_record = (timestamp[end_idx] - t_ref) / 1000.0
    
    label_start = "Record Start (Marqueur)" if i == 0 else None
    label_end = "Record End (Marqueur)" if i == 0 else None
    
    plt.axvline(x=t_start_record, color='black', linestyle='--', linewidth=1, label=label_start)
    plt.axvline(x=t_end_record, color='red', linestyle='--', linewidth=1, label=label_end)

plt.xlabel('Time (s) [relative to first record]')
plt.ylabel('Position Value (Centered)')
plt.title(f'Raw time series data (Segmented by {MARKER_FILE})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
